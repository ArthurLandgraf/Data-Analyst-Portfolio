The columns are currently following a **camelCase** format, and to ensure we have a better reading as well as align with most other Python projects, we will convert it to **snakecase**.

```python
def camel_to_snake(name):
  name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
  return re.sub('([a-z0-9])([A-Z])', r'\1_\2', name).lower()
```

```python
new_columns = []

for col in autos.columns:
    new_columns.append(camel_to_snake(col))

autos.columns = new_columns

autos.head()
```

## Brand Analysis
- Include steps of the process

```python
autos_final['brand'].value_counts()
```

Looking at the sheer number won't tell us much. So let's look at two specific pieces of information:
- Mean **price** per brand;
- Mean **odometer** value per brand.

And to ensure we get the most insightful analysis, we will consider only brands with more than 500 cars, and group the remaining ones all together.

```python
brand_mean_price = {}
brands = autos_final['brand'].unique()
brand_counts = autos_final['brand'].value_counts().to_dict()
brand_others = []

for brand in brands:
    brand_rows = autos_final['brand'] == brand
    mean_price = autos_final.loc[brand_rows, 'price_usd'].mean()
    if brand_counts[brand] > 500:
        brand_mean_price[brand] = mean_price
    else:
        brand_others.append(brand)

others_bool = autos_final['brand'] == brand_others[0]

for brand in brand_others[1:]:
    others_bool = others_bool | (autos_final['brand'] == brand)

brand_mean_price['others'] = autos_final.loc[others_bool , 'price_usd'].mean()
```

```python
df_price_means = pd.DataFrame.from_dict(data=brand_mean_price, orient='index', columns=['Mean Price'])
df_price_means.sort_values(by='Mean Price', ascending=False)
```

**Takeaways**:

- The difference between the TOP 4 and the next ones to come is pretty considerable;
- We will avoid consider 'Others' as it's grouped as it is exactly due to lack of data (brands with less than 500 cars);
- Audi, Mercedes Benz and BMW are a clear cut at the top;
- Skoda, Volkswagen and Toyota represents a middle ground.

Now let's look at the average odometer value per brand.

```python
brand_mean_odo = {}

for brand in brands:
    brand_rows = autos_final['brand'] == brand
    mean_odo = autos_final.loc[brand_rows, 'odometer_km'].mean()
    if brand_counts[brand] > 500:
        brand_mean_odo[brand] = mean_odo
    else:
        brand_others.append(brand)

brand_mean_odo['others'] = autos_final.loc[others_bool , 'odometer_km'].mean()
```

```python
df_odo_means = pd.DataFrame.from_dict(data=brand_mean_odo, orient='index', columns=['Mean Odometer'])
df_odo_means.sort_values(by='Mean Odometer', ascending=False)
```

**Takeaways**:

- BMW, Mercedes Benz and Audi figure in both Price and Odometer top ranking, meaning the aggregated value of the vehicle might not be correlated to the odometer value.
- The variance from TOP 1 to 7 is nearly 5000 km, which is not a very significant value.
- The lasgest cut actually comes from the bottom, in which Smart is more than 10 000 km a part from Skoda.

Now let's look more deeply into the correlation between both Price and Odometer.

```python
df_price_odo = pd.concat([df_price_means, df_odo_means], axis=1)
df_price_odo['km per usd'] = df_price_odo['Mean Odometer'] / df_price_odo['Mean Price']
df_price_odo.sort_values(by='km per usd', ascending=False)
```

The **'km per usd'** might seem difficult to understand, but think of it this way:
> If you could buy each car individual km registered in the odometer, the value listed there would indicate how many km you would be able to buy with 1 dollar.

And following this concept, if we take into consideration 2 well known facts:
- Cars with more km on the odometer tend to be worn out thus in worse condition
- The more worn out the car is, it's expected that it's price would go down

We can then assume that the higher **'km per usd'** is, better that brand would be. On the other hand, if the it's low it means your dollars are paying a high price regardless of km ran.

*Now let's see what is the most common model for each brand.*

```python
brand_common_model = {}

for brand in brands:
    brand_rows = autos_final['brand'] == brand
    common_model = autos_final.loc[brand_rows, 'model'].describe().top
    if brand_counts[brand] > 500:
        brand_common_model[brand] = common_model
    else:
        brand_others.append(brand)

brand_common_model['others'] = autos_final.loc[others_bool , 'model'].describe().top
```

```python
df_model = pd.DataFrame.from_dict(data=brand_common_model, orient='index', columns=['Common Model'])
df_brands = pd.concat([df_price_odo, df_model], axis=1)
df_brands
```

**Takeaways**:
